# Create Agents to Write a Child Book

```python
from crewai import Agent, Task, Crew
```

```python
# --- Agents ---

planner = Agent(
    role="Pete the Cat Story Architect",
    goal="Outline a repetitive story about {topic} using the 'step and change' formula.",
    backstory="You focus on the sequence of events: 1. New item, 2. Messy obstacle, "
              "3. Color change, 4. Resilience. You provide the raw beats for the writer.",
    allow_delegation=False,
    verbose=True
)

writer = Agent(
    role="Pete the Cat Narrator",
    goal="Write the story in clean, rhythmic prose with no headers or labels.",
    backstory="You write exactly like the source text. You use short sentences, "
              "repetitive songs, and the signature 'Did Pete cry? Goodness no!' "
              "rhetorical style. You output only the story text.",
    allow_delegation=False,
    verbose=True
)

editor = Agent(
    role="Pete the Cat Editor",
    goal="Remove all formatting, titles, and labels to leave only the story text.",
    backstory="You are a minimalist. You ensure the output is a clean .txt style "
              "document that starts immediately with the action and ends with the moral.",
    allow_delegation=False,
    verbose=True
)
```

```python
# --- Tasks ---

plan_task = Task(
    description=(
        "Create a plan for a story about {topic}. "
        "Identify: 1. The starting object (e.g., white shoes), "
        "2. Three obstacles (e.g., strawberries, mud, water), "
        "3. The resulting colors for each."
    ),
    expected_output="A simple list of the object, obstacles, and colors.",
    agent=planner,
)

write_task = Task(
    description=(
        "Write the story using the transcript style: "
        "Pete is walking -> Loves his item -> Sings song -> 'Oh no!' -> Steps in mess -> "
        "Color change -> 'Did Pete cry?' -> Singing continues. "
        "DO NOT include titles, subtitles, or 'Chapter' markers."
    ),
    expected_output="The full story text, structured with line breaks for the songs.",
    agent=writer,
)

edit_task = Task(
    description=(
        "Refine the story. Strip away any 'Introduction' or 'Final Answer' text. "
        "Ensure it matches the provided clean format: "
        "Starts with 'Pete the Cat was walking...' and ends with '...it's all good'."
    ),
    expected_output="A clean, plain-text story with no metadata or headers.",
    agent=editor
)
```

```python
# --- Crew ---

crew = Crew(
    agents=[planner, writer, editor],
    tasks=[plan_task, write_task, edit_task],
    verbose=2
)
```

```python
result = crew.kickoff(inputs={"topic": "Pete the Cat's trip in China"})
```
